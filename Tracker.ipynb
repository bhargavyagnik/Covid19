```python
import pygal
from pygal.style import Style
custom_style = Style(
  background='transparent',
  plot_background='transparent',colors=('#F7F7F7', '#E8537A', '#E95355', '#E87653', '#E89B53'))
def plot_date_graph(df,start_date,end_date,date_title,y_name,show_dots,path):
    daterange = pd.date_range(start_date, end_date)
    dates=[]
    cases=[]
    for single_date in daterange:
        d=single_date.strftime("%Y-%m-%d")
        dates.append(d)
        temp=df[df[date_title]==d]
        case=temp[y_name].values[0]
        cases.append(case)
    line_chart = pygal.Line(show_legend=False,show_y_labels=False,show_x_labels=False,style=custom_style)
    line_chart.add('Cases',cases, show_dots=show_dots)
    line_chart.render_to_file('static/'+path)
    
```
